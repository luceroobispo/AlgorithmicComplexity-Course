```python
# Pasar de una matriz de adyacencia a un grafo

import graphviz as gv

#Matriz de adyacencia dirigida

gv.Source("""
   digraph grafod1   {
     a -> {b,c};
     b -> c;
   }
""")
```

```python
#Probemos llevar una matriz de adyacencia a grafo (matriz A)

def dibujaGrafo(A):
  grafo = gv.Digraph("grafod2")
  n = len(A)

  for i in range(n):
    for j in range(n):
      if A[i,j] == 1:
        #print(f"Hay una arista desde el nodo {i+1} al nodo {j+1}")
        grafo.edge(str(i+1),str(j+1))

  return grafo
```

```python
migrafo = dibujaGrafo(A)

print(migrafo)
```

```python
#Matriz de Adyacencia No dirigida

#Matriz de adyacencia NO dirigida
gv.Source("""
   graph grafoNOd1   {
     a -- b;
     a -- c;
     c -- d;
     b -- d;
   }
""")
```

```python
#Probemos llevar una matriz de adyacencia NO DIRIGIDA a un grafo (matriz A)

def dibujaGrafoND(A):
  grafo = gv.Graph("grafoNOd1")
  n = len(A)

  for i in range(n):
    for j in range(n):
      if A[i,j] == 1:
        #print(f"Hay una arista desde el nodo {i+1} al nodo {j+1}")
        grafo.edge(str(i+1),str(j+1),)

  return grafo
```

```python
migrafoND = dibujaGrafoND(A)
```
