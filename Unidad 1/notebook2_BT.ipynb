### Combinaciones de sumar un numero

```python
# backtracking

def cmbSuma(n, i):
    cola = []
    sum = 0
    
    if(i==n):                   # caso base si n es 1
        return 1
    
    if(n%i==0):
        while(sum<n):
            sum = sum + i
            cola.append(i)
        print(cola)
        cola =[]
    
    else:
        j = i
        temp = 0
        while(sum<n and j>0):
            temp = sum + j
            if(temp<=n):
                cola.append(j)
                sum = temp
            else:    
                j = j-1
                
        print(cola)
        cola =[]
            
    cmbSuma(n, i+1)
    
    
cmbSuma(4, 1)
```

```python
# backtracking simplificado

def backtrack(n, cola:list , sum, i):
    if(sum >=n or i<1):               #funcion parada
        return
    
    temp = sum + i
    
    if(temp<=n):                #verificar suma
        cola.append(i)
        sum = temp
    else:
        i = i-1
        
    backtrack(n, cola, sum, i)  #llamado recursivo
    
    return cola

def cmbSuma(n):
    cola = []
    sum = 0
    i = 1
    while(i<n):
        cola = backtrack(n, cola, sum, i)
            
        if(len(cola)!=0):
            print(cola)
            cola = [] 
        i+=1
        
    
cmbSuma(8)
```

### AZUL, NEGRO, ROJO


```python
def backtracking(nivel, n, cola, pila):
  #Condicion de Fin
  if(nivel==n):
    print("pila", pila)
  else:
    j = nivel
    while( j < n ):
      # RESTRICCION
      if (not((nivel == 1 and cola[0] == "a") or (nivel == 0 and cola[0] == "r"))): # Si no aplica la Restriccion
        # cola.pop(0): saca el primer elmento de la cola
        pila.append(cola.pop(0))
           
        backtracking(nivel+1, n, cola, pila)
        
        cola.append(pila.pop())
        
      else: # Si aplica la Restriccion
        cola.append(cola[0])
        cola.pop(0)
      
      j = j + 1


def main():
  cola = ["a", "n", "r"]  # Lista de elementos a permutar # a:azul, n:negro, r:rojo
  pila = []               # Lista para mostrar las permutaciones 
  
  nivel = 0
  n = len(cola)

  backtracking(nivel, n, cola, pila)


main()
```

### Combinaciones de un rango de [1,n] agrupos en k

```python
def combinacion():
    n = 4
    k = 2
    ans = backtrack(n,k, 1, [], [])
    print(ans)
    
def backtrack(n, k, start, ans, cmb):
    if(len(cmb)==k): 
        ans.append(cmb.copy())
        
    
    for i in range(start, n+1):
        cmb.append(i)
        backtrack(n,k, i+1, ans, cmb)
        cmb.pop()
        
    return ans
    
    
    
combinacion()
```

## Sum
 Input: candidates = [2,3,6,7], target = 7
Output: [[2,2,3],[7]
Explanation:
2 and 3 are candidates, and 2 + 2 + 3 = 7. Note that 2 can be used multiple times.
7 is a candidate, and 7 = 7.
These are the only two combinations.

```python
def combinacion():
    target = 7
    lst = [2,3,6,7]
    ans = backtrack(target, lst, 0, 0, [], [])
    print(ans)
    
def backtrack(target, lst, start, sum, ans, cmb):
    if(sum == target): 
        ans.append(cmb.copy())
    
    for i in range(start, len(lst)):
        
        temp = sum + lst[i]
        
        
        if(temp<target):
            sum = temp
            cmb.append(lst[i]) #2
            backtrack(target, lst, start, sum, ans, cmb)
        else:
            if(lst[0] + sum <=target):
                backtrack(target, lst, start-1, sum, ans, cmb)
            else:
                cmb.clear()
                sum =0
                backtrack(target, lst, start+1, sum, ans, cmb)
            
            
        
        cmb.clear()
        
    return ans
    
    
    
combinacion()
```

### El numero es capicua

```python
## FUERZA BRUTA

def capicua(lst, i, d):
    band = True
    while(i<d):
        
        print(lst[i], " = ", lst[d])
        
        if(lst[i] != lst[d]):
            band = False
            return band
            
        
        i = i + 1
        d = d - 1
        
    return band
        

arr = [9,8 , 7, 8,9]
capicua(arr, 0, len(arr)-1)
```

```python
## BACKTRACKING

def capicua(lst, i, d, band):
    
    if(i>=d):
        return band
    else:

        print(lst[i], " = ", lst[d])
    
        if(lst[i] != lst[d]):
            band = False
    
        return capicua(lst, i+1, d-1, band)
        

arr = [9,8 , 6, 8,9]
if(capicua(arr, 0, len(arr)-1, True) == True):
    print("Arreglo Capicua")
else:
    print("Arreglo No Capicua")
```
