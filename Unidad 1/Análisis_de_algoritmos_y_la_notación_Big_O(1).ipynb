## 1. Introducción

**Problema:** Cálculo del factorial de un numero

A continuación desarrollaremos dos algoritmos que resuelvan el problema de cálculo de factorial del un numero.

**Algoritmo #1**

```python
def fact(n):
    producto = 1
    for i in range(n):
        producto = producto * (i+1)
    return producto

print(fact(5))
```

El algoritmo simplemente toma un número entero como argumento. Dentro de la funcion fact, una variable llamada producto se inicializa en 1. Un ciclo se ejecuta de 1 a **n** y durante cada iteración, el valor en producto se multiplica por el número que itera el ciclo y el resultado se almacena nuevamente en la variable producto. Después de que se ejecute el ciclo, la variable producto contendrá el factorial.

**Algorimo #2**

```python
def fact2(n):
    if n == 0:
        return 1
    else:
        return n * fact2(n-1)

print(fact2(5))
```

En este segunto algoritmo se utilizó una función recursiva para calcular el factorial de un numero.

#### Cuál de los dos algoritmos seria mejor utilizar?

Para obtener la respuesta, debemos encontrar la **complejidad del algoritmo**

Una forma de hacerlo es encontrando el tiempo requerido al ejecutar cada uno de los dos algoritmos (pero ya sabemos que no es la forma optima).

Usaremos el literal **%timeit** seguido de la llamada a la función para encontrar el tiempo que tarda cada una de las funciones en ejecutarse. 

```python
%timeit fact(50)
%timeit fact2(50)
```

El tiempo de ejecución muestra que el primer algoritmo es más rápido en comparación con el segundo algoritmo que involucra recursividad. Este ejemplo muestra la importancia del análisis de algoritmos. En el caso de entradas grandes, la diferencia de rendimiento puede volverse más significativa.

Sin embargo, ya sabemos que **el tiempo de ejecución no es una buena métrica para medir la complejidad de un algoritmo**, ya que depende del hardware (y otras consideraciones). Se necesita una métrica de análisis de complejidad más objetiva para los algoritmos. Aquí es donde entra en juego la **notación Big O**.

### 2. Análisis de algoritmos con notación Big-O

La **notación Big-O** es una métrica utilizada para encontrar la complejidad del algoritmo. 

Básicamente, la notación Big-O significa la relación entre la entrada al algoritmo y los pasos necesarios para ejecutar el algoritmo. 

Se denota con una gran "O" seguida de paréntesis de apertura y cierre. 
Dentro del paréntesis, va la relación entre la entrada y los pasos tomados por el algoritmo y se presenta usando "n".

#### 2.1. Orden de Complejidad Constante O(c)

Decimos que la complejidad de un algoritmo es constante si los pasos necesarios para completar su ejecución permanecen constantes, independientemente del número de entradas. La complejidad constante se denota por **O(c)** donde c puede ser cualquier número constante.

Escribamos un algoritmo simple en Python que encuentre el cuadrado del primer elemento de la lista y luego lo imprima en la pantalla.

```python
#O(5) -> O(1)

def algo_constante(items):
    resultado = items[0] * items[0] # 2 accesos + 1 operacion + 1 asignacion= 4
    print(resultado) # 1

algo_constante([4, 5, 6, 8])

# asignacion = 1 | edad = 8
# operacion matematica = 1  | a * b
# imprimir variable = 1 | print("xd")
# acceso = 1 | arreglo[1]
```

En el script anterior, independientemente del tamaño de entrada o la cantidad de elementos en la lista de entrada items, el algoritmo realiza solo 2 pasos: 
1. Encontrar el cuadrado del primer elemento
2. imprimir el resultado en la pantalla. 

Por lo tanto, la complejidad permanece constante.

Si dibujamos un diagrama de líneas con el tamaño de la variable de entrada items en el eje x y el número de pasos en el eje y, obtendremos una línea recta. Para visualizar esto, ejecutemos el siguiente script:

```python
import matplotlib.pyplot as plt
import numpy as np

x = [2, 4, 6, 8, 10, 12]

y = [2, 2, 2, 2, 2, 2]

plt.plot(x, y, 'b')
plt.xlabel('Entradas')
plt.ylabel('Pasos')
plt.title('Orden de Complejidad Constante O(c)')
plt.show()
```

#### 2.2. Orden de Complejidad Lineal O(n)

Decimos que la complejidad de un algoritmo es lineal si los pasos necesarios para completar la ejecución de un algoritmo aumentan o disminuyen linealmente con el número de entradas. La complejidad lineal se denota por **O(n)**.

En este ejemplo, mostraremos todos los elementos de la lista:

```python
#O(n)
def algo_lineal(items):
    for item in items: #n
        print(item) # 1

algo_lineal([4, 5, 6, 8])
```

En el codigo ejemplo anterior, la complejidad de la función **algo_lineal** es lineal, ya que el número de iteraciones del ciclo for será igual al tamaño del arreglo de entrada llamada items. Por ejemplo, si hay 4 elementos en la lista items, el ciclo for se ejecutará 4 veces, y así sucesivamente.

A continuación se muestra el gráfico de orden de complejidad lineal. Lo generamos con las entradas en el eje x y el número de pasos en el eje y.

```python
import matplotlib.pyplot as plt
import numpy as np

x = [2, 4, 6, 8, 10, 12]

y = [2, 4, 6, 8, 10, 12]

plt.plot(x, y, 'b')
plt.xlabel('Entradas')
plt.ylabel('Pasos')
plt.title('Orden de Complejidad Lineal O(n)')
plt.show()
```

Es importante destacar que en el caso exista una gran cantidad de entradas, las constantes se vuelven insignificantes. Por ejemplo, revisemos siguiente script:

```python

# n*1 + n*1 = 2n
# se eliminan constantes = n
# O(n)

def algo_lineal_2(items):
    for item in items: # n
        print(item) # 1

    for item in items: # n
        print(item) # 1

algo_lineal_2([4, 5, 6, 8])
```

En el script anterior, hay dos bucles for que iteran sobre la lista de entrada items. Por lo tanto, **la complejidad del algoritmo se convierte en O(2n)**, sin embargo, en el caso de infinitos elementos en la lista de entrada, el doble de infinito sigue siendo igual a infinito, por lo que podemos ignorar la constante 2 (ya que en última instancia es insignificante) y la complejidad del algoritmo sigue siendo O(n).

Podemos verificar y visualizar esto aún más al trazar las entradas en el eje x y el número de pasos en el eje y como se muestra a continuación:

```python
import matplotlib.pyplot as plt
import numpy as np

x = [2, 4, 6, 8, 10, 12]

y = [4, 8, 12, 16, 20, 24]

plt.plot(x, y, 'b')
plt.xlabel('Entradas')
plt.ylabel('Pasos')
plt.title('Orden de Complejidad Lineal O(n)')
plt.show()
```

En el script anterior, puede ver claramente que **y=2n**, sin embargo, la salida es lineal al igual que cuando **y=n**, por tanto, podemos ignorar la constante 2 o cualquier otra constante.

#### 2.3. Orden de Complejidad Cuadrática O(n^2)

Se dice que la complejidad de un algoritmo es cuadrática cuando los pasos requeridos para ejecutar un algoritmo son una función cuadrática del número de elementos en la entrada. La complejidad cuadrática se denota como **O(n^2)**. 

Veamos en el siguiente ejemplo como es una función con complejidad cuadrática:

```python
def algo_cuadratico(items):
    for item in items:
        for item2 in items:
            print(item, ' ' ,item)

algo_cuadratico([4, 5, 6, 8])
```

En el script anterior, tenemos un ciclo externo que itera a través de todos los elementos de la lista de entrada y luego un ciclo interno anidado, que nuevamente itera a través de todos los elementos de la lista de entrada. El número total de pasos realizados es n * n, donde n es el número de elementos en la matriz de entrada.

El siguiente gráfico traza el número de entradas frente a los pasos para un algoritmo con complejidad cuadrática.

```python
def plot_algo_cuadratico(lista, clr):
    x = lista
    y = [ (i**2) for i in x]
    plt.plot(x, y, linestyle='-', color=clr)
```

```python
lista = [4, 5, 6, 8]

plot_algo_cuadratico(lista, 'b')
plt.xlabel('Entradas')
plt.ylabel('Pasos')
plt.title('Orden de Complejidad Cuadrática O(n^2)')
```

**(BONO)** El siguiente codigo muestra la representacion gráfica de una ecuacion polinómica.

```python
#Visualizamos una ecuacion polinomica del tipo y = ax2 + bx + c
def plot_ecuacion_polinom(a, b, c, clr):
    #x es una lista de numeros entre -10 y 10
    x = list(range(-10, 11))
    #construimos la ecuacion donde cada item de x esta representado por i
    y = [ (a*i**2 + b*i + c) for i in x]
    plt.plot(x, y, linestyle='-', color=clr)
```

```python
# Inicializamos los datos de nuestra ecuacion polinomica
a = -1
b = 0
c = 0

plot_ecuacion_polinom(a, b, c, 'g')
plt.xlabel('Entradas')
plt.ylabel('Pasos')
plt.title('Ecuación Polinomica')
```

### 3. Encontrar la complejidad algorítmica de funciones complejas

En los ejemplos anteriores, hemos hemos desarrollado solo una función en la entrada. ¿Qué sucede si se realizan varias funciones en la entrada? Desarrollemos el siguiente ejemplo.

```python
def algo_complejo(items):

    for i in range(5):
        print("Me encanta el curso de Complejidad Algoritmica")

    for item in items:
        print(item)

    for item in items:
        print(item)

    print("Big O")
    print("Big O")
    print("Big O")

algo_complejo([4, 5, 6, 8])
```

En el código anterior, se realizan varias tareas:

1. Se imprime una cadena 5 veces en la consola usando la declaración print, 
2. Imprimimos la lista de entrada dos veces en la pantalla y, 
3. Finalmente, se imprime otra cadena tres veces. 

Para encontrar la complejidad de dicho algoritmo, necesitamos dividir el código del algoritmo en partes e intentar encontrar la complejidad de las piezas individuales.

**1. Se imprime una cadena 5 veces en la consola usando la declaración print,**

    for i in range(5):
        print("Me encanta el curso de Complejidad Algoritmica")
        
   La complejidad de esta parte es **O(5)**. Dado que se están realizando cinco pasos constantes en esta pieza de código, independientemente de la entrada.
    

**2. Imprimimos la lista de entrada dos veces en la pantalla **
    for item in items:
        print(item)
 
    for item in items:
        print(item)
        
   Sabemos que la complejidad del código anterior es **O(n)** que resulta de O(n) + O(n) = O(2n) = O(n) porque eliminamos el valor cosntante 2.
    

**3. Se imprime otra cadena tres veces.**

    print("Big O")
    
    print("Big O")
    
    print("Big O")

Finalmente, una cadena se imprime tres veces, por lo que la complejidad es **O(3)**

Ahora corresponde sumar las ordenes de complejidad algoritmica calculadas parcialmente:

**O(5) + O(n) + O(n) + O(3)**

Simplificando lo anterior obtenemos:

**O(8) + O(2n) = O(n)**

Eliminamos O(8) y la constante 2 del orden de comlejidad O(2n) porque dijimos anteriormente que cuando la entrada (que tiene una longitud n en este caso) se vuelve extremadamente grande, las constantes se vuelven insignificantes, es decir, el doble o la mitad del infinito sigue siendo infinito. 

Por lo tanto, podemos ignorar las constantes. 

La complejidad final del algoritmo será **O(n)**.

### Complejidad algoritmica del peor caso vs. el mejor caso

Por lo general, cuando se nos pregunta sobre la complejidad del algoritmo, debemos calcularla el peor de los casos. 

Para comprender la complejidad del mejor caso y del peor caso, revisemos el siguiente código:

```python
def algo_buscar(num, items):
    for item in items:          #n = n(1 + 1 + 1) = 3n
        if item == num:         #1
            return True         #1
        else:
            pass                #1
nums = [2, 4, 6, 8, 10]         

print(algo_buscar(2, nums))
```

En el código anterior, tenemos una función que toma un número y una lista de números como entrada. 
- Devuelve verdadero si el número pasado se encuentra en la lista de números, de lo contrario, devuelve None. 
- Si busca 2 en la lista, se encontrará en la primera comparación. Esta es la complejidad del mejor de los casos del algoritmo en el que el elemento buscado se encuentra en el primer índice buscado. 
- La complejidad del mejor caso, en este caso, es O(1). 
- Por otro lado, si busca 10, se encontrará en el último índice buscado. 
- El algoritmo tendrá que buscar en todos los elementos de la lista, por lo que la complejidad del peor de los casos se convierte en O(n).

### 4. Complejidad Espacial

Además de la complejidad del tiempo, donde cuenta la cantidad de pasos necesarios para completar la ejecución de un algoritmo, también puede encontrar la complejidad del espacio, que se refiere a la cantidad de espacios que necesita asignar en el espacio de memoria durante la ejecución de un programa.

Veamos el siguiente ejemplo:

```python
def retorna_cuadrados(n):
    lista_cuadrados = []
    for num in n:
        lista_cuadrados.append(num * num)

    return lista_cuadrados

nums = [2, 4, 6, 8, 10]
print(retorna_cuadrados(nums))
```

- En el código anterior, la función acepta una lista de enteros y devuelve una lista con los cuadrados de enteros correspondientes. 
- El algoritmo tiene que asignar memoria para el mismo número de elementos que en la lista de entrada. 
- Por lo tanto, la complejidad espacial del algoritmo se convierte en **O(n)**.
