# **Problema de la Mochila - Enfoque Programación Dinámica**

```python
'''
Capacidad de la mochila = W
Lista de pesos : wt = []
Lista de precios o ganancia : pr = []
No. de productos = N
Tabla de programacion dinamica: dp = []
'''

# Funcion para calcular el Beneficio Máximo
def kProfit(W,N,wt,pr,dp):
    # Condicion Base: (Tamaño de la Lista == 0) o (Capacidad de la Mochila == 0) 
    if N==0 or W==0: #Condicion Base de la Recursividad
        return 0
    # Si el subproblema se resolvió previamente, lo devolvemos
    if dp[N][W] is not None:
        return dp[N][W]
    if wt[N-1] <= W:  #Si peso[N-1] < Capacidad_Mochila
        dp[N][W] = max(pr[N-1]+kProfit(W-wt[N-1],N-1,wt,pr,dp), kProfit(W,N-1,wt,pr,dp))
        return dp[N][W]
    else: ##Si peso[N-1] >= Capacidad_Mochila
        dp[N][W] = kProfit(W,N-1,wt,pr,dp)

        #print("dp= ", dp)
        return dp[N][W]
```

```python
if __name__ == '__main__':
  # Ejemplo de PPT
  #W = 11                    #Capacidad de la Mochila
  #wt = [1, 2, 5, 6, 7]      #Dimensión|Pesos
  #pr = [1, 6, 18, 22, 28]   #Ganancia

  # Ejemplo de Esquema de Aprendizaje 
  W = 8                         #Capacidad de la Mochila
  wt = [1, 2, 4, 5, 7, 8]       #Dimensión|Pesos
  pr = [2, 5, 6, 10, 13, 16]    #Ganancia
  
  N = len(pr) # N=6
  # Definir matriz DP | Matriz de Productos x Dimensiones(Pesos)
  dp = [[None] * (W + 1) for _ in range(N + 1)] #Genera la Matrix (N+1)x(W+1) rellena con None: Matriz de 7x9 
  #print("dp = ", dp)

  # Llame a kProfit para calcular el beneficio máximo
  maxProfit = kProfit(W,N,wt,pr,dp)
  
  print('Maximo beneficio es : ', maxProfit)
```
