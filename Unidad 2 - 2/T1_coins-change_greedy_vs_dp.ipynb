# **Problema del Cambio de Monedas**

## **Enfoque Codicioso**
Creamos una funcion llamada greedycoins, que recibe un conjunto de denominaciones de distintas monedas (ordenadas de menor a mayor denominacion) y el importe o resultado objetivo.

```python
# pt: precio total - resultado objetivo
# d: arreglo de monedas posibles

def greedycoins(pt, d):
    monedas = []
    i = len(d) - 1              # empieza a recorre el ciclo desde el la moneda con mas valor (ENFOQUE CODICCIOSO)
    
    while pt > 0:
        while d[i] > pt:        # hacer que el precio total sea mayor que la monedas posible
            i -= 1
        monedas.append(d[i])    # agregar la moneda de mas valor a la respuesta, siempre y cuando sea menor que el precio objetivo
        pt -= d[i]              # actualizar el precio total
        
    return monedas
```

Probamos la funcion greedycoins con dos importes distintos.

```python
d = [1, 5, 10, 20, 50]
print(greedycoins(40, d))
print(greedycoins(85, d))
```

Probamos la funcion greedycoins con un diferente conjunto de monedas y dos importes distintos.

```python
d = [1, 5, 10, 20, 25, 50]
print(greedycoins(40, d))
print(greedycoins(85, d))
```

# Enfoque segun Programación Dinámica (DP Dynamic Programming)

```python
# pt: precio total - resultado objetivo
# d: arreglo de monedas posibles

def dpcoins(pt, d):
    C = [0]*(pt + 1)
    S = [0]*(pt + 1)
    
    for pi in range(1, pt + 1):
        mind = 0
        for k in range(len(d)):
            if d[k] > pi: break
            if C[pi - d[k]] < C[pi - d[mind]]:
                mind = k
        C[pi] = C[pi - d[mind]] + 1
        S[pi] = d[mind]

    coins = []
    while pt > 0:
        coins.append(S[pt])
        pt -= S[pt]
    
    return coins
```

```python
d = [1, 5, 10, 20, 50]
print(dpcoins(40, d))
print(dpcoins(85, d))
```
