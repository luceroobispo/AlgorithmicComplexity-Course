```python
# REF: http://bryukh.com/labyrinth-algorithms/

from collections import deque

def maze2graph(maze):
    height = len(maze)  #Heigh of maze
    width = len(maze[0]) if height else 0 #Width of maze
    graph = {(i, j): [] for j in range(width) for i in range(height) if not maze[i][j]} #Create the dictionary where key=(i,j); value=element
    for row, col in graph.keys():
        if row < height - 1 and not maze[row + 1][col]:
            graph[(row, col)].append(("S", (row + 1, col)))
            graph[(row + 1, col)].append(("N", (row, col)))
        if col < width - 1 and not maze[row][col + 1]:
            graph[(row, col)].append(("E", (row, col + 1)))
            graph[(row, col + 1)].append(("W", (row, col)))
    return graph
```

```python
#Using Breadth First Search (BFS)
def find_path_bfs(maze):
    start, goal = (1, 1), (len(maze) - 2, len(maze[0]) - 2) #(Begin, End)
    queue = deque([("", start)])  #The QUEUE have the first element: ("", start)
    visited = set() #The visited
    graph = maze2graph(maze)
    while queue:  #Using Queue for BFS
        path, current = queue.popleft()
        if current == goal:
            return path
        if current in visited:
            continue
        visited.add(current)
        for direction, neighbour in graph[current]:
            queue.append((path + direction, neighbour))
    return "NO WAY!"
```
